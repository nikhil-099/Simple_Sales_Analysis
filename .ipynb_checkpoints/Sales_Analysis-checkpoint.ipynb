### Task1: Read and Merge or Concatnate datasets into a new file 

```python
files = [file for file in os.listdir('./Sales_data')]
#Creating an empty Data frame 
new_datas = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_data/"+file)
    new_datas = pd.concat([new_datas, df])
    
new_datas.to_csv("all_data.csv", index = False)
```

## Reading in Updated data 

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

## Cleaning up Data 

### Droping NaN values

```python
Nan_df = all_data[all_data.isna().any(axis = 1)] ##Filter to see the Nan values in data frame form
Nan_df
all_data = all_data.dropna(how="all")
```

```python
all_data.head()
```

## Finding 'Or' and removing it

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or'] ##Filtering data not having "or" value
```

### Augmenting Months Column and Changing Data types

```python
all_data['Months'] = all_data['Order Date'].str[0:2]
all_data['Months'] = all_data['Months'].astype('int32')
all_data.head()
```

# Q1: What was the best month for sales? How much was earned that month

## Creating Sales column 

```python
#Before creating sales column we need to convert data types of 'Quantity Ordered' and 'Price Each'
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
#Creating Sales Column
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
results = all_data.groupby('Months').sum()
```

```python
import matplotlib.pyplot as plt
months = range(1, 13)

plt.bar(months, results['Sales'])
plt.title("Sales Per Month")
plt.ylabel("Sales in USD (in Million)")
plt.xlabel('Month Number')
plt.xticks(months)
plt.show()
```

# Q2: What US city had the highest number of Sales

```python
##Function for get city part of address
def get_city(address):
    return address.split(',')[1]

##Function for get state part of address
def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x:f"{get_city(x)}({get_state(x)})")
all_data
```

```python
results = all_data.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt

##As the order of city column in "results" and "all_data" dataframe are different so we used list comprehension to match the order
cities = [city for city, df in all_data.groupby('City')]  
plt.barh(cities, results['Sales'])
plt.title("Sales Per City")
plt.xlabel("Sales in USD (in Million)")
plt.ylabel('Month Number')
#plt.xticks(cities)
plt.show()
```

# Q3: What time should we display ad to maximize likeliehood of customer's buying product?

```python
all_data["Order Date"] = pd.to_datetime(all_data["Order Date"])
```

```python
all_data["Hour"] = all_data["Order Date"].dt.hour
all_data['Minute'] = all_data["Order Date"].dt.minute
```

```python
import matplotlib.pyplot as plt

hours = [hour for hour, df in all_data.groupby("Hour")]  
plt.plot(hours, all_data.groupby("Hour").count(), c = "#8a2be2")
plt.title("Advertisement Running")
plt.xlabel("Hour")
plt.ylabel("No. Of Orders Made")
plt.xticks(hours)
plt.grid()
plt.show()
```

# Q4: What products are most often sold together

```python
## Making demo dataframe with duplicate "Order ID"
df = all_data[all_data["Order ID"].duplicated(keep=False)]

## Creating a separate column for the products sold together as per same Product ID
## This will throw a warning "SettingWithCopyWarning" as there are two parts which perform different task 
df["Grouped"] = df.groupby("Order ID")["Product"].transform(lambda x: ",".join(x))
df.head()
```

```python
## We will drop the duplicate rows
df = df[["Order ID", "Grouped"]].drop_duplicates()
df
```

```python
## To iterate and keep the count of product pairs we need import below imported modules
from itertools import combinations
from collections import Counter

## To keep count
count = Counter()

for row in df["Grouped"]:
    row_list = row.split(",")
    ## It will count the number combination of a two or three products (Depends on parameter in combinations method)
    count.update(Counter(combinations(row_list, 2))) 

for key, value in count.most_common(20):
    print(key, value)
```

# Q5: What Product sold the most? Why do you think it sold the most?

```python
product_group = all_data.groupby("Product") ##Groupby products
quantity_ordered = product_group.sum()["Quantity Ordered"] ## Find sum of quantities for a specific product 

products = [products for products, df in product_group] 

plt.bar(products, quantity_ordered)
plt.xlabel("Quantity Ordered")
plt.ylabel("Products")
plt.xticks(products, rotation = 'vertical')
plt.grid(axis="y")
plt.show()
```

```python
prices = all_data.groupby("Product").mean()["Price Each"]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx() ##Create a twin Axes sharing the xaxis.
ax1.bar(products, quantity_ordered, color = "#0dd0b4")
ax2.plot(products, prices, color = "#101010")

ax1.set_xlabel("Product Name")
ax1.set_ylabel("Quantity Ordered", color = "#0dd0b4")
ax2.set_ylabel("Price($)", color = "#101010")
ax1.set_xticklabels(products, rotation='vertical')

plt.show()
```
